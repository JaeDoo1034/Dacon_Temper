```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
```

###  Data 구분 : Y00~ Y17 / Y18
- 30일간의 데이터(Y00~Y17) : Data_T
- 3일간의 데이터(Y18) : Data_E

```python
Data.shape
```

##### 30일간 데이터 전처리

```python
Data_T = Data.iloc[:4320,:]
```

```python
Data_T.shape
```

```python
Data_T.drop(['Y18'],axis = 1,inplace = True)
```

```python
Data_T = Data_T.set_index('id')
```

```python
Data_T.head()
```

```python
Data_T.to_csv('data/Data_30.csv')
```

##### 3일간 데이터 전처리

```python
Data_E = Data[Data['Y18'] > 0]
```

```python
Data_E.shape
```

```python
Data_E = Data_E.drop(['Y00', 'Y01', 'Y02', 'Y03', 'Y04', 'Y05', 'Y06', 'Y07', 'Y08', 'Y09',
       'Y10', 'Y11', 'Y12', 'Y13', 'Y14', 'Y15', 'Y16', 'Y17'],axis = 1)
```

```python
Data_E = Data_E.set_index('id')
```

```python
Data_E.head()
```

```python
Data_E.to_csv('data/Data_3.csv')
```

#### Y(종속변수) 데이터

##### 30일

```python
Label_T = ["Y" +'0' + str(i) if i < 10 else "Y" + str(i) for i in range(18) ]
```

```python
Data_T_Y = Data.loc[:,Label_T]
```

```python
Data_T_Y = Data_T_Y.dropna(axis = 0)
```

```python
with open('data/Data_T_Y.pk','wb') as f:
    pickle.dump(Data_T_Y,f)
```

##### 3일

```python
Data_E_Y = Data.loc[:,'Y18']
```

```python
Data_E_Y = Data_E_Y.dropna(axis = 0)
```

```python
with open('data/Data_E_Y.pk','wb') as f:
    pickle.dump(Data_E_Y,f)
```
